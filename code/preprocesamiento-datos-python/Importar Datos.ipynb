```python
import pandas as pd

df = pd.read_csv('Data/datos_exportados.csv')
```

```python
df['P_1'] = pd.Categorical(df['P_1'], categories=['Hombre', 'Mujer'], ordered=True)
df['P_2'] = pd.Categorical(df['P_2'], categories=['18 a 25 años', '26 a 35 años', '36 a 45 años', '45 a 55 años', '56 a 65 años', '66 años en adelante'], ordered=True)
df['P_3'] = pd.Categorical(df['P_3'], categories=['1 a 5 años', '6 a 10 años', '11 años o más'], ordered=True)
df['P_4'] = pd.Categorical(df['P_4'], categories=['Norte', 'Centro', 'Sur'], ordered=True)

df['VD_34'] = pd.Categorical(df['VD_34'], categories=['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo'], ordered=True)
df['VD_35'] = pd.Categorical(df['VD_35'], categories=['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo'], ordered=True)
df['VD_36'] = pd.Categorical(df['VD_36'], categories=['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo'], ordered=True)
df['VD_37'] = pd.Categorical(df['VD_37'], categories=['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo'], ordered=True)
df['VD_38'] = pd.Categorical(df['VD_38'], categories=['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo'], ordered=True)
```

## Variables Independientes

```python
likert = ['Totalmente en desacuerdo', 'Desacuerdo', 'Indiferente', 'De acuerdo', 'Totalmente de acuerdo']
dispositivos = ['Computadora', 'Teléfono celular inteligente (Smartphone)', 'Fax', 'Tablets', 'Webcam', 'Otro']
recurso_tecnologico = ['Redes sociales', 'Correos electrónicos', 'Llamadas celulares', 'Mensajería instantánea (WhatsApp,Telegram)', 'Videoconferencia (Zoom, Meet, Webinar, WhatsApp, Telegram)', 'Otro', 'Ninguno']

def vi_01_antes(elementos):
    VI_01_antes = ''
    
    if elementos.VI_14a == 1:
        VI_01_antes = likert[4]
    elif elementos.VI_15a == 1:
        VI_01_antes = likert[3]
    elif elementos.VI_16a == 1:
        VI_01_antes = likert[2]
    elif elementos.VI_17a == 1:
        VI_01_antes = likert[1]
    elif elementos.VI_18a == 1:
        VI_01_antes = likert[0]
        
    return VI_01_antes

def vi_01_despues(elementos):
    VI_01_despues = ''
    
    if elementos.VI_14b == 1:
        VI_01_despues = likert[4]
    elif elementos.VI_15b == 1:
        VI_01_despues = likert[3]
    elif elementos.VI_16b == 1:
        VI_01_despues = likert[2]
    elif elementos.VI_17b == 1:
        VI_01_despues = likert[1]
    elif elementos.VI_18b == 1:
        VI_01_despues = likert[0]

    return VI_01_despues

def vi_02_antes(elementos):
    VI_02_antes = ''
    
    if elementos.VI_19a == 1:
        VI_02_antes = dispositivos[0]
    elif elementos.VI_20a == 1:
        VI_02_antes = dispositivos[1]
    elif elementos.VI_21a == 1:
        VI_02_antes = dispositivos[2]
    elif elementos.VI_22a == 1:
        VI_02_antes = dispositivos[3]
    elif elementos.VI_23a == 1:
        VI_02_antes = dispositivos[4]
    elif elementos.VI_24a == 1:
        VI_02_antes = dispositivos[5]
    
    return VI_02_antes

def vi_02_despues(elementos):
    VI_02_despues = ''
    
    if elementos.VI_19b == 1:
        VI_02_despues = dispositivos[0]
    elif elementos.VI_20b == 1:
        VI_02_despues = dispositivos[1]
    elif elementos.VI_21b == 1:
        VI_02_despues = dispositivos[2]
    elif elementos.VI_22b == 1:
        VI_02_despues = dispositivos[3]
    elif elementos.VI_23b == 1:
        VI_02_despues = dispositivos[4]
    elif elementos.VI_24b == 1:
        VI_02_despues = dispositivos[5]
    
    return VI_02_despues

def vi_03_antes(elementos):
    VI_03_antes = ''
    
    if elementos.VI_25a == 1:
        VI_03_antes = recurso_tecnologico[0]
    elif elementos.VI_26a == 1:
        VI_03_antes = recurso_tecnologico[1]
    elif elementos.VI_27a == 1:
        VI_03_antes = recurso_tecnologico[2]
    elif elementos.VI_28a == 1:
        VI_03_antes = recurso_tecnologico[3]
    elif elementos.VI_29a == 1:
        VI_03_antes = recurso_tecnologico[4]
    elif elementos.VI_30a == 1:
        VI_03_antes = recurso_tecnologico[5]
    elif elementos.VI_31a == 1:
        VI_03_antes = recurso_tecnologico[6]
        
    return VI_03_antes

def vi_03_despues(elementos):
    VI_03_despues = ''
        
    if elementos.VI_25b == 1:
        VI_03_despues = recurso_tecnologico[0]
    elif elementos.VI_26b == 1:
        VI_03_despues = recurso_tecnologico[1]
    elif elementos.VI_27b == 1:
        VI_03_despues = recurso_tecnologico[2]
    elif elementos.VI_28b == 1:
        VI_03_despues = recurso_tecnologico[3]
    elif elementos.VI_29b == 1:
        VI_03_despues = recurso_tecnologico[4]
    elif elementos.VI_30b == 1:
        VI_03_despues = recurso_tecnologico[5]
    elif elementos.VI_31b == 1:
        VI_03_despues = recurso_tecnologico[6]
    
    return VI_03_despues
```

## Variables Dependientes

```python
num_clientes = ['5 a 10 clientes', '11 a 15 clientes', '16 a 20 clientes', '21 clientes o más']
facturacion_mensual = ['Menor a $5.000', '$5.000 a $10.000', '$10.000 a $50000', '$50.000 a $100.000', 'Mayores a $100.000']
pagos_servicios = ['Efectivo', 'Tarjeta de crédito/débito', 'Transferencias', 'Paypal']
metodo_facturacion = ['Facturas físicas', 'Facturas electrónicas']

def vd_01_antes(elementos):
    VD_01_antes = ''
        
    if elementos.VD_5a == 1:
        VD_01_antes = num_clientes[0]
    elif elementos.VD_6a == 1:
        VD_01_antes = num_clientes[1]
    elif elementos.VD_7a == 1:
        VD_01_antes = num_clientes[2]
    elif elementos.VD_8a == 1:
        VD_01_antes = num_clientes[3]
    
    return VD_01_antes

def vd_01_despues(elementos):
    VD_01_despues = ''
        
    if elementos.VD_5b == 1:
        VD_01_despues = num_clientes[0]
    elif elementos.VD_6b == 1:
        VD_01_despues = num_clientes[1]
    elif elementos.VD_7b == 1:
        VD_01_despues = num_clientes[2]
    elif elementos.VD_8b == 1:
        VD_01_despues = num_clientes[3]
    
    return VD_01_antes

def vd_02_antes(elementos):
    VD_02_antes = ''
        
    if elementos.VD_9a == 1:
        VD_02_antes = facturacion_mensual[0]
    elif elementos.VD_10a == 1:
        VD_02_antes = facturacion_mensual[1]
    elif elementos.VD_11a == 1:
        VD_02_antes = facturacion_mensual[2]
    elif elementos.VD_12a == 1:
        VD_02_antes = facturacion_mensual[3]
    elif elementos.VD_13a == 1:
        VD_02_antes = facturacion_mensual[4]
    
    return VD_01_antes

def vd_02_despues(elementos):
    VD_02_despues = ''
        
    if elementos.VD_9b == 1:
        VD_02_despues = facturacion_mensual[0]
    elif elementos.VD_10b == 1:
        VD_02_despues = facturacion_mensual[1]
    elif elementos.VD_11b == 1:
        VD_02_despues = facturacion_mensual[2]
    elif elementos.VD_12b == 1:
        VD_02_despues = facturacion_mensual[3]
    elif elementos.VD_13b == 1:
        VD_02_despues = facturacion_mensual[4]
    
    return VD_02_despues

def vd_03(elementos):
    VD_03 = ''
        
    if elementos.VD_32a_ == 1:
        VD_03 = pagos_servicios[0]
    elif elementos.VD_32b_ == 1:
        VD_03 = pagos_servicios[1]
    elif elementos.VD_32c_ == 1:
        VD_03 = pagos_servicios[2]
    elif elementos.VD_32d_ == 1:
        VD_03 = pagos_servicios[3]
    
    return VD_03

def vd_04(elementos):
    VD_04 = ''
        
    if elementos.VD_33a_ == 1:
        VD_04 = metodo_facturacion[0]
    elif elementos.VD_33b_ == 1:
        VD_04 = metodo_facturacion[1]
    
    return VD_04
```

## Depuración de los datos

```python
datos_nombres = ['P_1', 'P_2', 'P_3', 'P_4', 'VI_01_antes', 'VI_01_despues', 'VI_02_antes', 'VI_02_despues', 'VI_03_antes', 'VI_03_despues', 
                 'VD_01_antes', 'VD_01_despues', 'VD_02_antes', 'VD_02_despues', 
                 'VD_03', 'VD_04', 'VD_05', 'VD_06', 'VD_07', 'VD_08', 'VD_09']
datos = []

for elementos in df.itertuples():
    VI_01_antes = vi_01_antes( elementos )
    VI_01_despues = vi_01_despues( elementos )
    VI_02_antes = vi_02_antes( elementos )
    VI_02_despues = vi_02_despues( elementos )
    VI_03_antes = vi_03_antes( elementos )
    VI_03_despues = vi_03_despues( elementos )
    
    VD_01_antes = vd_01_antes( elementos )
    VD_01_despues = vd_01_despues( elementos )
    VD_02_antes = vd_02_antes( elementos )
    VD_02_despues = vd_02_despues( elementos )
    VD_03 = vd_03( elementos )
    VD_04 = vd_04( elementos )
    
    dato = [ elementos.P_1, elementos.P_2, elementos.P_3, elementos.P_4,  
             VI_01_antes, VI_01_despues, VI_02_antes, VI_02_despues, VI_03_antes, VI_03_despues, 
             VD_01_antes, VD_01_despues, VD_02_antes, VD_02_despues, VD_03, VD_04,
             elementos.VD_34, elementos.VD_35, elementos.VD_36, elementos.VD_37, elementos.VD_38 ]
    
    datos.append(dato)
             
datos_finales = pd.DataFrame( datos, columns=datos_nombres )
```

Modificación de las variables a categóricas.

```python
datos_finales['VI_01_antes'] = pd.Categorical(datos_finales['VI_01_antes'], categories=likert, ordered=True)
datos_finales['VI_01_despues'] = pd.Categorical(datos_finales['VI_01_despues'], categories=likert, ordered=True)

datos_finales['VI_02_antes'] = pd.Categorical(datos_finales['VI_02_antes'], categories=dispositivos, ordered=True)
datos_finales['VI_02_despues'] = pd.Categorical(datos_finales['VI_02_despues'], categories=dispositivos, ordered=True)

datos_finales['VI_03_antes'] = pd.Categorical(datos_finales['VI_03_antes'], categories=recurso_tecnologico, ordered=True)
datos_finales['VI_03_despues'] = pd.Categorical(datos_finales['VI_03_despues'], categories=recurso_tecnologico, ordered=True)
```

```python
datos_finales['VD_01_antes'] = pd.Categorical(datos_finales['VD_01_antes'], categories=num_clientes, ordered=True)
datos_finales['VD_01_despues'] = pd.Categorical(datos_finales['VD_01_despues'], categories=num_clientes, ordered=True)

datos_finales['VD_02_antes'] = pd.Categorical(datos_finales['VD_02_antes'], categories=facturacion_mensual, ordered=True)
datos_finales['VD_02_despues'] = pd.Categorical(datos_finales['VD_02_despues'], categories=facturacion_mensual, ordered=True)

datos_finales['VD_03'] = pd.Categorical(datos_finales['VD_03'], categories=pagos_servicios, ordered=True)
datos_finales['VD_04'] = pd.Categorical(datos_finales['VD_04'], categories=metodo_facturacion, ordered=True)
```
